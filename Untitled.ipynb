```python
total_yr['gender']= total_yr['gender'].astype(str)
```

```python
total_yr.dtypes
```

```python
#change numerical values to male, female, not specified
not_spec = total_yr.replace(to_replace ="0", 
                 value ="not specified") 

male = not_spec.replace(to_replace ="1", 
               value ="male")

gender = male.replace(to_replace ="2", 
                  value ="female") 

# total_yr.replace(to_replace ="1", 
#                  value ="male") 

# total_yr.replace(to_replace ="2", 
#                  value ="female") 

gender
```

```python
# change usertype

subscriber= gender.replace(to_replace ="Subscriber", 
                value ="Annual Member")

user_type = subscriber.replace(to_replace ="Customer", 
                  value ="24 Hr / 3 Day Pass") 

user_type
```

```python
#export as CSV
total_yr.to_csv('2020.csv', index=False)
```
